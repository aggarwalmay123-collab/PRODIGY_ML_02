# ***Create a K-means clustering algorithm to group customers of a retail store based on their purchase history.***


From the given dataset we choose only two columns in order to get a refined 2D plot . The two columns that I would be choosing are:

1.   Annual Income(k$)
2.   Spending Score(1-100)







# Step 1:  Importing the Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Step 2 : Importing the Dataset

```python
dataset=pd.read_csv('Mall_Customers.csv')
X=dataset.iloc[:,[3,4]].values
```

# Step 3 : Using the Elbow Method to find the optimal number of clusters

```python
from sklearn.cluster import KMeans
# Now we make a list of "within Cluster Sum of Squares"
wcss=[]
# Now we make a for loop for 10 clusters
for i in range(1,11):
  # we create an object for KMeans
  kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
  #k-means++helps us to avoid Random Initialization
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)
  #.inertia_ helps us to gibe wcss its desired squared value
#Now we would plot a graph between "Number Of Clusters" and "WCSS"
plt.plot(range(1,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number Of Clusters')
plt.ylabel('WCSS')
plt.show()
```

# Step 4 : Training the K-Means model on the dataset

```python
#From the graph we conclude that the Nmber of Clusters are 5
kmeans=KMeans(n_clusters=5,init='k-means++',random_state=42)
#y_kmeans identifies the 5 clusters
y_kmeans=kmeans.fit_predict(X)
```

```python
print(y_kmeans)
```

# Step 5 : Visualising the Clusters

```python
#For the FIRST CLUSTER
plt.scatter(X[y_kmeans==0,0], X[y_kmeans==0,1] , s=100 , c='red', label='Cluster 1')
#For the SECOND CLUSTER
plt.scatter(X[y_kmeans==1,0], X[y_kmeans==1,1] , s=100 , c='blue', label='Cluster 2')
#For the THIRD CLUSTER
plt.scatter(X[y_kmeans==2,0], X[y_kmeans==2,1] , s=100 , c='green', label='Cluster 3')
#For the FOURTH CLUSTER
plt.scatter(X[y_kmeans==3,0], X[y_kmeans==3,1] , s=100 , c='pink', label='Cluster 4')
#For the FIFTH CLUSTER
plt.scatter(X[y_kmeans==4,0], X[y_kmeans==4,1] , s=100 , c='magenta', label='Cluster 5')
#CENTROIDS
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1] , s=300, c='yellow', label ='Centroids')
#Plotting the graph between "Annual Income(k$)" and "Spending Score(1-100)"
plt.title('Cluster Of Customers')
plt.xlabel('Annual Income(k$)')
plt.ylabel('Spending Score(1-100)')
plt.legend()
plt.show()
```
